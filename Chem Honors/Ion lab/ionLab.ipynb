## Results

```python
from matplotlib import patches
import matplotlib.pyplot as plt
from matplotlib.table import Table

# presets / constants
ppt = 'ppt'
noppt = ''
colLabels = ['Sodium Hydroxide', 'Sodium Phosphate', 'Sodium Carbonate', 'Sodium Iodide', 'Sodium Sulfate']
rowLabels = ['Silver Nitrate', 'Lead II Nitrate', 'Cobalt Nitrate', 'Copper II Nitrate', 'Barium Nitrate', 'Iron II Nitrate', 'Magnesium Nitrate']


expePrediction = [
    [ppt, ppt, ppt, ppt, noppt],
    [ppt, ppt, ppt, ppt, ppt],
    [ppt, ppt, ppt, noppt, noppt],
    [ppt, ppt, ppt, noppt, noppt],
    [noppt, ppt, ppt, noppt, ppt],
    [ppt, ppt, ppt, noppt, noppt],
    [ppt, ppt, ppt, noppt, noppt],
]
expeResult = [
    [ppt, ppt, ppt, ppt, noppt],
    [noppt, noppt, ppt, ppt, ppt],
    [ppt, ppt, ppt, noppt, noppt],
    [ppt, ppt, ppt, ppt, noppt],
    [ppt, ppt, ppt, noppt, ppt],
    [ppt, ppt, ppt, noppt, noppt],
    [ppt, noppt, ppt, noppt, noppt],
]


def ionicLab(experiment_prediction:list , experiment_result: list):
    fig, axs = plt.subplots(ncols=2, figsize=(12, 6))
    plt.box(on=None)
    # plt.axis('off')
    fig.subplots_adjust(right=1.5)
    # set u the axes 
    for ax in axs:
        ax.set_xlim([0, 50])
        ax.set_ylim([0, 50])
        ax.axis('tight')
        ax.grid(False)
        ax.set_axis_off()

    colorScheme = []
    # calculate disagreements 
    totNumberOfExpe = 0
    totAgreements = 0
    # check for disagreements
    for r in range(len(expePrediction)):
        colorSchemeRow = []
        expePredictionRow = expePrediction[r]
        expeResultRow = expeResult[r]
        for c in range(len(expePredictionRow)):
            totNumberOfExpe += 1
            if expePredictionRow[c] != expeResultRow[c]:
                colorSchemeRow.append('#fc1c1c') # red
                continue
            totAgreements += 1
            colorSchemeRow.append('#fff') #clear
        colorScheme.append(colorSchemeRow)
    
    print(totAgreements)
    
    # predictions 
    ax=axs[0]
    prediction = ax.table(
        cellText=experiment_prediction,
        cellColours=colorScheme,
        colLabels=colLabels,
        rowLabels=rowLabels,
        loc='upper left',
    )
    prediction.auto_set_font_size(False)
    prediction.auto_set_column_width(col=list(range(8)))
    ax.set_title('Prediction ', loc='center')

    prediction.scale(2, 3)

    # results 
    ax = axs[1]
    result = ax.table(
        cellText=experiment_result,
        cellColours=colorScheme,
        colLabels=colLabels,
        rowLabels=rowLabels,
        loc='upper left',
    )
    result.auto_set_font_size(False)
    result.auto_set_column_width(col=list(range(8)))
    ax.set_title('Result ', loc='center')

    result.scale(2, 3)

    # legends
    ax = axs[0]
    patch1 = patches.Patch(
        color='red', label='Disagreements between Prediction and Result')
    patch2 = patches.Patch(
        color='#ffff', label='# of Agreement:' + str(totAgreements) + '\t % of Agreement = ' + str(int((totAgreements/totNumberOfExpe) * 100)) + r'$\ $' + '\tThe Solubility rules are mostly reliable. \t')
    ax.legend(
        handles=[patch1, patch2],
        ncol=2,
        loc='lower left',
        mode="expand",
        bbox_transform=fig.transFigure, 
        bbox_to_anchor=(0, 0.11, 1.488, 1),
        fontsize=14,
        )

ionicLab(expePrediction, expeResult)

```
