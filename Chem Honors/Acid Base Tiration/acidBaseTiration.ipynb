### Data 

$\#g$ container + NaOH: $70.0g$

$V_{NaOH}$ solution $= 100ml$

### Calculations for [NaOH]
$$\begin{align*}
[NaOH] &= \frac{n_{NaOH}}{V} \\
       &= \frac{2.41g\ NaOH}{100ml} * \frac{10^3}{L} * \frac{1 mol\ NaOH}{40g\ NaOH} \\
       &= 0.6025\ M\ NaOH

\end{align*}$$

##### Titration curve(left) and First derivative(right) of data:

```python
import numpy as np
import matplotlib.pyplot as plt  # plotting & visuals
from IPython.display import Markdown as md # allowing markdown to be written in python and use python variables 

# data
vbase = [ 2, 4, 6, 8, 10, 12, 13, 13.5, 14, 15, 16, 17] 
ph = [1.2, 1.3, 1.4, 1.6, 1.8, 2.1, 2.3, 7.1, 9.7, 10, 10.1, 10.2]
slope = []

# 0 for padding
vbase.append(0)
ph.append(0)
for i in range(len(vbase) - 1):
    slope.append((ph[i] - ph[i + 1]) / (vbase[i] - vbase[i + 1]))

vbase.pop()
ph.pop()

fig = plt.figure('main', figsize=(12, 5))

phVsBaseGraph = fig.add_subplot(121)
slopeGraph = fig.add_subplot(122)

axis = plt.gca()

phVsBaseGraph.plot(vbase, ph)
phVsBaseGraph.set_xlabel('$V_{base}(ml)$')
phVsBaseGraph.set_ylabel("pH")
phVsBaseGraph.set_title('ph vs $V_{base}$')


slopeGraph.plot(vbase, slope)
slopeGraph.set_xlabel('$V_{base}(ml)$')
slopeGraph.set_ylabel('dP/dVh')
slopeGraph.set_title('dP/dVh vs. $V_{base}(ml)$')

plt.show()

```

### Calculations of [HCl] based off endpoint and equivalence point
$$\begin{align*}
a\ [A]\ V_{A} &= b\ [B]\ V_{B} \\
[HCl]\ V_{HCl} &= [NaOH]\ V_{NaOH}
\end{align*}$$
##### ***@end point***:
$$\begin{align*}
[HCl]\ 30ml &= 0.6025\ M * 13.5\ ml \\
      [HCl] &= \frac{0.6025\ M * 13.5\ ml}{30\ ml} \\
      [HCl] &= 0.271\ M
\end{align*}$$

##### ***@equivalence point***:
$$\begin{align*}
[HCl]\ 30ml &= 0.6025\ M * 13.5\ ml \\
      [HCl] &= \frac{0.6025\ M * 13.5\ ml}{30\ ml} \\
      [HCl] &= 0.271\ M
\end{align*}$$




### Conclusion

The concentration of [HCl] obtained from either the visual end point or the equivalence point is the same. Meaning that if recorded correctly, both methods will yield the same results.
